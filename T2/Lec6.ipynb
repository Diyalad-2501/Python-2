```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
from sklearn.tree import DecisionTreeClassifier , plot_tree
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , mean_absolute_error , r2_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
import plotly.express as px
import plotly.io as pio
pio.renderers.default='notebook'
```

```python
df = pd.read_csv('diabetes.csv')
df = df.dropna()
```

```python
x = df.drop(['Outcome'] , axis=1) 
y = df['Outcome']

x_train , x_test , y_train , y_test = train_test_split( x , y , test_size=0.2 , random_state=42)
from sklearn.svm import SVC
svm = SVC(C=1)
svm.fit(x_train , y_train)
y_pred = svm.predict(x_test)
print(classification_report(y_test , y_pred))
```

```python
df = pd.read_csv('Teen_Mental_Health_Dataset.csv')
```

```python
df.head()
```

```python
x=pd.get_dummies(df.drop(['depression_label'],axis=1))
y=df['depression_label']
t=[0.05, 0.10, 0.15, 0.20, 0.25]
d = {"model":[] , "acc":[] , "test":[]}
for i in t:
    x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=i , random_state=42)
    
    d['model'].append('knn')
    d['test'].append(i)
    knn = KNeighborsClassifier(n_neighbors=29)
    knn.fit(x_train,y_train)
    d['acc'].append(accuracy_score(y_test , knn.predict(x_test)))

    d['model'].append('DecisionTree')
    d['test'].append(i)
    dt=DecisionTreeClassifier(criterion='entropy' , max_depth=4 , random_state=42)
    dt.fit(x_train , y_train)
    d['acc'].append(accuracy_score(y_test , dt.predict(x_test)))

    d['model'].append('Random Forest Classifier')
    d['test'].append(i)
    rfc = RandomForestClassifier(n_estimators=100 , max_depth=4 , random_state=42)
    rfc.fit(x_train , y_train)
    d['acc'].append(accuracy_score(y_test , rfc.predict(x_test)))

    d['model'].append('Support Vector Machine')
    d['test'].append(i)
    svm = SVC(C=10 , kernel='rbf' , random_state=42)
    svm.fit(x_train,y_train)
    d['acc'].append(accuracy_score(y_test , svm.predict(x_test)))
new = pd.DataFrame(d)
print(new)
```

```python
px.line(data_frame=new, x='test' ,y='acc' , color='model')
```

```python
result = new.groupby('model')['acc'].max().reset_index()
px.bar(data_frame=result , x='model',y='acc')
```

```python
from sklearn.preprocessing import PolynomialFeatures
df = pd.read_csv('created_house_price_prediction.csv')
df = df.dropna()
x = pd.get_dummies(df.drop(['Price'] , axis=1))
y= df['Price']

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2 , random_state=42)
mse=[]
d={'degree':[] , 'r2':[]}

for i in range(1,6):
    d['degree'].append(i)
    poly = PolynomialFeatures(degree=i)
    x_train_poly = poly.fit_transform(x_train)
    x_test_poly = poly.transform(x_test)

    model = LinearRegression()
    model.fit(x_train_poly , y_train)
    y_pred = model.predict(x_test_poly)
    
    mse.append(mean_squared_error(y_test , y_pred))
    d['r2'].append(r2_score(y_test , y_pred))
d= pd.DataFrame(d)
m = d['r2'].max()
print(d[d['r2']==m]['degree'])
```

```python
test_size=[0.05, 0.10, 0.15, 0.20, 0.25]
d={'model':[],'test_size':[] , 'r2':[],"mse":[]}

for i in test_size:
    x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=i , random_state=42)
    d['test_size'].append(i)
    d['model'].append('Linear Regression')
    model = LinearRegression()
    model.fit(x_train , y_train)
    y_pred = model.predict(x_test)
    d['mse'].append(mean_squared_error(y_test , y_pred))
    d['r2'].append(r2_score(y_test , y_pred))

    d['test_size'].append(i)
    d['model'].append('Polynomial Regression')
    poly = PolynomialFeatures(degree=5)
    x_train_poly = poly.fit_transform(x_train)
    x_test_poly = poly.transform(x_test)

    model = LinearRegression()
    model.fit(x_train_poly , y_train)
    y_pred = model.predict(x_test_poly)
    
    d['mse'].append(mean_squared_error(y_test , y_pred))
    d['r2'].append(r2_score(y_test , y_pred))
d = pd.DataFrame(d)
```
