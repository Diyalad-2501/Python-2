```python
df["Total_Rowby"]=df.apply(sum,axis=1)
```

```python
def grade(x):
    if x>300:
        return "A+"
    if x>250:
        return "A"
    if x>200:
        return "B"
    return "F"
```

```python
df["Grade"]=df["Total_Rowby"].apply(grade)
```

```python
df
```

```python
df.apply(lambda x:"Fail" if((x["A"]<35) or (x["B"]<35) or (x["C"]<35) or( x["D"]<35) or (x["E"]<35)) else "Pass",axis=1)
```

```python
def res(x):
    if x["A"]<35 or x["B"]<35 or x["C"]<35 or x["D"]<35 or x["E"]<35:
        return "Fail"
    return "Pass"
```

```python
df.apply(res,axis=1)
```

```python
def bonus(y):
    return y+2
```

```python
df["A"]=df["A"].apply(bonus)
```
