```python
def likesGB(x):
    if x>2000:
        return "Good"
    return"Bad"
```

```python
df["GB"]=df["likes"].apply(likesGB)
```

```python
df
```

```python
def risk_level(x):
    if x["likes"]>500 and x["comments"]<100:
        return "High"
    if x["likes"]>2000:
        return "Medium"
    if x["shares"]>1000:
        return "Low"
```

```python
df["risk_level"]=df.apply(risk_level,axis=1)
```
