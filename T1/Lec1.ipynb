```python
import pandas as pd
import numpy as np
```

```python
ds = pd.Series([10,20,30,40])
print(ds)
```

```python
d = pd.Series((10,20,30,40))
print(d)
```

```python
ds = pd.Series((10,20,30,40),index=['A','B','C','D'])
print(ds)
```

```python
d =pd.Series({'d':25,'i':1,'y':2007,'a':"hii"})
print(d)
```

```python
d = pd.Series(np.array([10,20,30,40]))
print(d)
print(type(d))
```

```python
# for 2 dimensional 
# Dataframe is a class of panda
d = pd.DataFrame(np.array([[10,20],[30,40],[50,60]]))
print(d)
```

```python
d = pd.DataFrame(np.array([[10,20],[30,40],[50,60]]) , index=['A','B','C'])
print(d)
```

```python
d = pd.DataFrame(np.array([[10,20],[30,40],[50,60]]) , index=['A','B','C'],columns=['P','Q'])
print(d)
```

```python
# for dictonary key as column 
d = pd.DataFrame({'d':[10,20],'i':[30,40],'y':[50,60]})
print(d)
```

```python
d = pd.DataFrame({'d':[10,20],'i':[30,40],'y':[50,60]},index=['A','B'])
print(d)
```

```python
d = pd.DataFrame({'d':[10,20],'i':[30,40],'y':[50,60]},index=['A','B'],columns=['P','Q','R'])
print(d)
```

```python
# in list of dictionary , dictionary key as column name and values in row
d = pd.DataFrame([{'d':1,'i':2},{'d':3,'i':4}])
print(d)
```

```python
d = pd.DataFrame([{'d':1,'i':2},{'d':3,'i':4}],index=['a','b'])
print(d)
```
