```python
df1 = pd.DataFrame({'name':['vha' , 'psp' , 'vby' , 'mss' , 'dps'],
                    'sub':['python','fsd','coa','dm','toc'],
                     'id':[3,2,1,4,6]})

df2 = pd.DataFrame({'id':[1,2,3,4,5] ,
                   'mark':[9,8,10,8,8]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1 , df2 , on='id')
```

```python
pd.merge(df1 , df2 , on='id' ,how ='inner')
```

```python
pd.merge(df1 , df2 , on='id' ,how ='outer')
```

```python
pd.merge(df1 , df2 , on='id' ,how ='left')
```

```python
pd.merge(df1 , df2 , on='id' ,how ='right')
```

```python
pd.merge(df1 , df2 ,how ='cross')
```

```python
pd.merge(df1 , df2 ,on='id',how ='cross')
```

```python
pd.merge(df1 , df2 ,left_index = True ,  right_index =True ,suffixes=('_df1','_df2'))
```

```python
df1 = pd.DataFrame({'name':['vha' , 'psp' , 'vby' , 'mss' , 'dps'],
                    'sub':['python','fsd','coa','dm','toc'],
                     'in':[3,2,1,4,6]})

df2 = pd.DataFrame({'id':[1,2,3,4,5] ,
                   'mark':[9,8,10,8,8]})
```

```python
df1.merge(df2 , left_on='in' , right_on ='id')
```

```python
df1 = pd.DataFrame({'name':['vha' , 'psp' , 'vby' , 'mss' , 'dps'],
                    'sub':['python','fsd','coa','dm','toc'],
                     'in':[3,2,1,4,6]})

df2 = pd.DataFrame({'id':[1,2,3,4,5] ,
                   'mark':[9,8,10,8,8],
                   'fac':['vha' , 'nas' , 'vby' ,'mss','dps']})
```

```python
# if colums name are different in both dataframe
df1.merge(df2 , left_on='name' , right_on ='fac' , how='outer')
```
